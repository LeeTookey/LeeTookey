# Activation Functions with Pytorch



```python
# Importing Required Libraries
import torch
import matplotlib.pyplot as plt


# Linear Activation Function
def linear(x):
    '''Returns input as it is'''
    return x

# Sigmoid Activation Function
def sigmoid(x):
    '''Returns 1 / (1 + exp(-x))'''
    return torch.sigmoid(x)

# Hyperbolic Tangent Activation Function
def tanh(x):
    '''Returns (exp(x) - exp(-x)) / (exp(x) + exp(-x))'''
    return torch.tanh(x)

# Rectified Linear Unit Activation Function
def relu(x):
    '''Returns x if x > 0, else 0'''
    return torch.relu(x)

# Leaky ReLU Activation Function
def leaky_relu(x, alpha=0.01):
    '''Returns x if x > 0, else alpha*x'''
    return torch.nn.functional.leaky_relu(x, alpha)

# Softmax Activation Function
def softmax(x):
    '''Compute softmax values for each sets of scores in x.'''
    return torch.nn.functional.softmax(x, dim=0)

# Test input
#x = torch.tensor([-0.25, 0.1, 0.2, 0.3])
x = torch.arange(-1 , 1, 0.1)

print("Linear:", linear(x))
print("Sigmoid:", sigmoid(x))
print("Tanh:", tanh(x))
print("ReLU:", relu(x))
print("Leaky ReLU:", leaky_relu(x))
print("Softmax:", softmax(x))

# Plot Graphs

# Create a figure and axis
#fig, ax = plt.subplots()

#y_list = [[linear(x)], [sigmoid(x)], [tanh(x)], [relu(x)], [leaky_relu(x)], [softmax(x)]]

# Loop over each y series in the list
#for i, y in enumerate(y_list):
#    # Create a new figure for each series
#    plt.figure(i)

    # Plot the data
 #   plt.plot(x, y)

    # Show the plot
#   plt.show()
# Plotting
# Plotting each activation function separately
plt.figure(figsize=(10, 6))
plt.plot(x.numpy(), linear(x).numpy())
plt.title('Linear Activation Function')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(x.numpy(), sigmoid(x).numpy())
plt.title('Sigmoid Activation Function')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(x.numpy(), tanh(x).numpy())
plt.title('Tanh Activation Function')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(x.numpy(), relu(x).numpy())
plt.title('ReLU Activation Function')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(x.numpy(), leaky_relu(x).numpy())
plt.title('Leaky ReLU Activation Function')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(x.numpy(), softmax(x).numpy())
plt.title('Softmax Activation Function')
plt.grid(True)
plt.show()

# Plotting all activation functions together
plt.figure(figsize=(10, 6))
plt.plot(x.numpy(), linear(x).numpy(), label='Linear')
plt.plot(x.numpy(), sigmoid(x).numpy(), label='Sigmoid')
plt.plot(x.numpy(), tanh(x).numpy(), label='Tanh')
plt.plot(x.numpy(), relu(x).numpy(), label='ReLU')
plt.plot(x.numpy(), leaky_relu(x).numpy(), label='Leaky ReLU')
plt.plot(x.numpy(), softmax(x).numpy(), label='Softmax')
plt.legend()
plt.grid(True)
plt.title('All Activation Functions')
plt.show()
#This code defines each activation function and then applies them to a range of test input values between -1 and 1 inclusive in steps of 0.1. It then plots the output of each activation function separately and together on a single graph. You can replace the test input with your own data. Remember to use appropriate activation functions based on the specific requirements of your model and the problem youre trying to solve. Happy coding! 

```
